```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
import rainflow as rf # https://github.com/iamlikeme/rainflow
```

```python
def rf_array2df(signal):

    df_rf = pd.DataFrame([
        [rng, mean, count, i_start, i_end] 
        for rng, mean, count, i_start, i_end in rf.extract_cycles(signal)
    ],columns=['rng', 'mean', 'count', 'i_start', 'i_end'])

    df_rf['vmin'] = df_rf['mean']-df_rf['rng']/2
    df_rf['vmax'] = df_rf['mean']+df_rf['rng']/2
    df_rf['N'] = df_rf['count']
    df_rf = df_rf.drop(['rng','mean','count'],axis=1)
    
    return df_rf
```

```python
signal1 = [0,100,50,200,40,300,-100,0]
signal2 = [0,-20,100,20,120,-400,-20,-100,10,0]
```

```python
rf1 = rf_array2df(signal1)
rf1
```

```python
rf2 = rf_array2df(signal2)
rf2
```

```python
rf12 = rf_array2df(signal1+signal2)
rf12
```
